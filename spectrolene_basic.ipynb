```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
def chromaplot(filename,song):
    ##read file
    Fs, sr = librosa.load("data/"+filename)
    ##stft on .wav
    chroma = librosa.feature.chroma_stft(y = Fs, sr = sr)
    
    ##plotting
    fig, ax = plt.subplots()
    fig.set_figheight(5)
    fig.set_figwidth(7.5)
    img = librosa.display.specshow(chroma,y_axis='chroma',x_axis='time',ax = ax)
    ax.set(title = song)
    fig.colorbar(img,ax=ax)
    
    
    
def melplot(filename,song,start_time,end_time):
    ##read file
    Fs, sr = librosa.load("data/"+filename)
    ##melspectrogram on .wav
    mel = librosa.feature.melspectrogram(y = Fs[start_time*sr:end_time*sr], sr = sr, n_mels = 128,fmax=8000)
    mel_dB = librosa.power_to_db(mel, ref = np.max)
    
    ##plotting
    fig, ax = plt.subplots()
    fig.set_figheight(5)
    fig.set_figwidth(7.5)
    img = librosa.display.specshow(mel_dB,y_axis='mel',x_axis='time',fmax = 8000, ax = ax)
    ax.set(title = song+": "+str(start_time)+"s to "+str(end_time)+"s.")
    fig.colorbar(img,ax=ax,format='%+2.0f dB')
    
def longmelplot(filename,song,start_time,end_time):
    ##read file
    Fs, sr = librosa.load("data/"+filename)
    ##melspectrogram on .wav
    mel = librosa.feature.melspectrogram(y = Fs[start_time*sr:end_time*sr], sr = sr, n_mels = 128,fmax=8000)
    mel_dB = librosa.power_to_db(mel, ref = np.max)
    
    ##plotting
    fig, ax = plt.subplots()
    fig.set_figheight(5)
    fig.set_figwidth(12.5)
    img = librosa.display.specshow(mel_dB,y_axis='mel',x_axis='time',fmax = 8000, ax = ax)
    ax.set(title = song+": "+str(start_time)+"s to "+str(end_time)+"s.")
    fig.colorbar(img,ax=ax,format='%+2.0f dB')
```
