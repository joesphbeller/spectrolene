```python
#setting correct environment
from pydub import AudioSegment
import numpy as np
from scipy.io import wavfile
from plotly.offline import init_notebook_mode
import plotly.graph_objs as go
import plotly
from plotly.offline import iplot
import plotly.io as pio
plotly.offline.init_notebook_mode(connected = True)
import matplotlib.pyplot as plt
```

```python
##This uses plotly and will generate an interactive graph. For waveforms, this is not needed, and it so far cannot do audio very long. It also will be timely.
def plotly_wave_form(file_name):
    fs_wav, data_wav = wavfile.read("data/" + file_name)
    duration = data_wav.shape[0]/fs_wav
    
    
    mins = int(duration/60)
    secs = int(duration - mins*60)
    print("Signal duration is ", mins, "minutes and", secs,"seconds.")
    check = input("Please make sure this is correct with the audio file (y/n?)") # <- Check 2
    
    
    
    if check == "y":
        #next, must normalize:
        data_wav_norm = data_wav / (2**15) #2^15 resolution of this file is 16 bits per sample
        time_wav = np.arange(0, len(data_wav)) / fs_wav
        #cropping and / 2 for longer samples
        crop = 60
        crop1 = data_wav_norm[0 * fs_wav: crop * fs_wav]
        crop2 = data_wav_norm[crop * fs_wav:]
        #plotting
        trace0 = go.Scatter(
            x = time_wav[0:crop*fs_wav],
            y = crop1)
        trace1 = go.Scatter(
            x = time_wav[crop*fs_wav:],
            y = crop2)
    
        plotly.offline.iplot([trace0,trace1])
    
    else: 
        print("Please, put either 'y' or 'n' into the prompted input, or try again with a different file")
```

```python
##This uses matplotlib-- which is much faster, and works with longer lengths
def plt_wave_form(file_name,title):
    fs_wav, data_wav = wavfile.read("data/" + file_name)
    duration = data_wav.shape[0]/fs_wav
    print(np.size(data_wav))
    
    mins = int(duration/60)
    secs = int(duration - mins*60)
    print("Signal duration is ", mins, "minutes and", secs,"seconds.")
    check = input("Please make sure this is correct with the audio file (y/n?)") # <- Check 2
    if check == "y":
        #next, must normalize:
        data_wav_norm = data_wav / (2**15) #2^15 resolution of this file is 16 bits per sample
        time_wav = np.arange(0, len(data_wav)) / fs_wav
        
        plt.plot(time_wav,data_wav_norm,color='pink')
        plt.xlabel("Time")
        plt.ylabel("Amplitude")
        plt.title(title)
        
    else: 
        print("Please, put either 'y' or 'n' into the prompted input, or try again with a different file")
```
